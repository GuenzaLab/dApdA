```python
import numpy as np
import matplotlib.pyplot as plt
import physt
```

```python
distance = np.load('Distance.npy')
twist = np.load('Twist.npy')
tilt = np.load('Tilt.npy')
roll = np.load('Roll.npy')
```

```python
for i in range(1,6):
    print(i)
    frames = np.loadtxt('frames_'+str(i),dtype=int)-1
    
    print(np.mean(distance[frames]))
    print(np.std(distance[frames]))
    print("\n")
    print(np.var(distance[frames]))

    print("\n")
```

```python
#Naive mean and variance for the twist angles within each macrostate

for i in range(1,6):
    print(i)
    frames = np.loadtxt('frames_'+str(i),dtype=int)-1
    
    print(np.rad2deg(np.mean(twist[frames])))
    #print(np.var(twist[frames]))
    print(np.rad2deg(np.var(twist[frames])))

    print("\n")
```

```python
for i in range(1,6):
    print(i)
    frames = np.loadtxt('frames_'+str(i),dtype=int)-1
    #print(frames)

    sp = 0
    cp = 0

    #Circular mean
    sp = np.array(np.sin(twist[frames]).sum())
    cp = np.array(np.cos(twist[frames]).sum())
    print(np.arctan2(sp,cp))
    print("\n")
    print(np.rad2deg(np.arctan2(sp,cp)))

    #Circular variance
    var = 1-(np.sqrt(np.sin(twist[frames]).sum()**2+np.cos(twist[frames]).sum()**2)/len(frames))
    print(var)

    print("\n")
    
    #physt.special.DirectionalHistogram(np.cos(twist[frames]),np.sin(twist[frames]))
```

```python
import scipy.stats

for i in range(1,6):
    print(i)
    frames = np.loadtxt('frames_'+str(i),dtype=int)-1
    #print(frames)
    print(scipy.stats.circmean(twist[frames],low=-np.pi,high=np.pi))
    print(scipy.stats.circvar(twist[frames],low=-np.pi,high=np.pi))

    print("\n")
```

```python
import astropy.stats

for i in range(1,6):
    print(i)
    frames = np.loadtxt('frames_'+str(i),dtype=int)-1
    #print(frames)
    print(astropy.stats.circstats.circmean(twist[frames]))
    print(astropy.stats.circstats.circvar(twist[frames]))

    print("\n")
```

```python
#Tilt

for i in range(1,6):
    print(i)
    frames = np.loadtxt('frames_'+str(i),dtype=int)-1
    #print(frames)

    print('Base 1')

    #Circular mean
    sp = np.array(np.sin(tilt[:,0][frames]).sum())
    cp = np.array(np.cos(tilt[:,0][frames]).sum())
    print(np.arctan2(sp,cp))

    #Circular variance
    var = 1-(np.sqrt(np.sin(tilt[:,0][frames]).sum()**2+np.cos(tilt[:,0][frames]).sum()**2)/len(frames))
    print(var)

    print("\n")
    
    print('Base 2')

    #Circular mean
    sp = np.array(np.sin(tilt[:,1][frames]).sum())
    cp = np.array(np.cos(tilt[:,1][frames]).sum())
    print(np.arctan2(sp,cp))

    #Circular variance
    var = 1-(np.sqrt(np.sin(tilt[:,1][frames]).sum()**2+np.cos(tilt[:,1][frames]).sum()**2)/len(frames))
    print(var)

    print("\n")
```
